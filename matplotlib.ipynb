```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle
import matplotlib.patches as patches


def create_profile_gui():
    fig, ax = plt.subplots(1, 1, figsize=(5, 7))

    # Синий фон вверху
    ax.add_patch(Rectangle((0, 0.7), 1, 0.3, color='lightblue', alpha=0.7))

    # Круг для фото профиля
    ax.add_patch(Circle((0.5, 0.85), 0.07, color='lightgreen', alpha=0.8))

    # Текст информации
    texts = [
        (0.5, 0.6, 'Иван Драго', 16, 'bold', 'center'),
        (0.05, 0.52, 'Биография', 14, 'bold', 'left'),
        (0.05, 0.47, 'Я инженер-программист с 10-летним', 10, 'normal', 'left'),
        (0.05, 0.44, 'опытом создания потрясающего кода.', 10, 'normal', 'left'),
        (0.05, 0.37, 'Умения', 14, 'bold', 'left'),
        (0.05, 0.33, 'Python | PHP | SQL | JavaScript', 10, 'normal', 'left'),
        (0.05, 0.27, 'Опыт', 14, 'bold', 'left'),
        (0.05, 0.23, 'Python Разработчик', 10, 'normal', 'left'),
        (0.05, 0.20, 'Март 2011 - настоящее время', 8, 'normal', 'left'),
        (0.05, 0.16, 'Водитель доставки пиццы', 10, 'normal', 'left'),
        (0.05, 0.13, 'Авг 2015 - Дек 2017', 8, 'normal', 'left')
    ]

    for x, y, text, size, weight, align in texts:
        ax.text(x, y, text, fontsize=size, fontweight=weight,
                ha=align, transform=ax.transAxes)

    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')
    ax.set_facecolor('white')

    plt.tight_layout()
    plt.show()


# Запускаем
create_profile_gui()
```
